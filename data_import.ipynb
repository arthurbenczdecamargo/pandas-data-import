```python
# SQLALCHEMY DATABASE
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
# SQL ENGINE
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url3 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
data = pd.read_csv(url3)
data.head()
```

```python
# CSV TO SQL
```

```python
data.to_sql('clients', engine, index=False)
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clients WHERE Categoria_de_renda = "Empregado"'
```

```python
employees = pd.read_sql(query, engine)
employees.head(10)
```

```python
employees.to_sql('employees', con=engine, index=False)
```

```python
pd.read_sql_table('employees', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual']).head(10)
```

```python
query = 'SELECT * FROM clients'
```

```python
pd.read_sql(query, engine).head(10)
```

```python
# DELETE FROM TABLE
```

```python
query = text("DELETE FROM clients WHERE ID_Cliente=5008804")
with engine.connect() as conn:
    conn.execute(query)
    conn.commit()
```

```python
pd.read_sql_table('clients', engine).head(10)
```

```python
# UPDATE TABLE
```

```python
query = text("UPDATE clients SET Grau_escolaridade = 'Ensino superior' WHERE ID_Cliente=5008806")
with engine.connect() as conn:
    conn.execute(query)
    conn.commit()
```

```python
pd.read_sql_table('clients', engine).head(10)
```
